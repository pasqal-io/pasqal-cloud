# Solving a Quadratic Unconstrained Binary Optimization (QUBO) problem on OVH

This notebook illustrates how to use Qubo Solver to solve QUBO problems using a QPU or emulator.

We begin by importing some relevant modules:

```python
%env PASQAL_PULSER_ACCESS_TOKEN=""

try:
    import torch
    import qoolqit
except ImportError:
    !pip install qubo-solver

# Force pasqal-cloud 0.23.0, we won't keep that but it's to have our new emulator backends available
!pip install pasqal-cloud==0.23.0

from qubosolver import QUBOInstance
from qubosolver.config import SolverConfig
from qubosolver.solver import QuboSolver
from qubosolver.config import QPU
from pasqal_cloud import OVHConnection

# define QUBO
Q = torch.tensor([[-0.2, 0, 1.0], [0, 0, 1.5], [1.0, 1.5, 0]])
instance = QUBOInstance(coefficients=Q)

connection = OVHConnection()

# Get available devices
print(f"Available devices: {connection.fetch_available_devices()}")
# Choose a device
device = qoolqit.Device.from_connection(connection, "FRESNEL_CAN1")
config = SolverConfig(
    use_quantum=True, device=device, backend=QPU(connection=connection, num_shots=1000)
)
# Run the solver
solver = QuboSolver(instance, config)
solutions = solver.solve()

# Display results
print(solutions)
```

#### On a remote emulators

Emulators are also available remotely via `pasqal_cloud`:

```python
from pasqal_cloud import RemoteMPSBackend
from qubosolver.backends import RemoteEmulator

# define QUBO
Q = torch.tensor([[-0.2, 0, 1.0], [0, 0, 1.5], [1.0, 1.5, 0]])
instance = QUBOInstance(coefficients=Q)

connection = OVHConnection()
config = SolverConfig(
    use_quantum=True,
    backend=RemoteEmulator(backend_type=RemoteMPSBackend, connection=connection),
)
# Run the solver
solver = QuboSolver(instance, config)
solutions = solver.solve()

# Display results
print(solutions)
```
